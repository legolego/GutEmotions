```python
import requests
import pandas as pd
```

```python
list_of_books = range(0, 1000)
lst_dicts = []

for book_id in list_of_books:
    base_url = "https://gutendex.com/books/{}".format(book_id)
    response = requests.get(base_url)
    print(response.json())
    if (response.status_code == 200) & ("authors" in response.json()):
        if (len(response.json()["authors"]) > 0):
            book_dict = {}
            book_dict["id"] = response.json()["id"]
            book_dict["title"] = response.json()["title"]
            book_dict["author"] = response.json()["authors"][0]["name"]
            book_dict["author_birth"] = response.json()["authors"][0]["birth_year"]
            book_dict["author_death"] = response.json()["authors"][0]["death_year"]
            book_dict["txt_link"] = "https://www.gutenberg.org/ebooks/{}.txt.utf-8".format(book_id)

            lst_dicts.append(book_dict)
        else:
            print("zero author length")
    else:
        print("No authors")

df = pd.DataFrame(lst_dicts)
df
```

```python
df['emotions'] = ""
df['emotion_scores'] = ""
df['summaries'] = ""

df.to_parquet("../streamlit/data/gutenberg_books.parquet", index=False)
df
```
